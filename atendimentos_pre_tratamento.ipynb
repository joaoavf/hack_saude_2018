```python
import pandas as pd
```

```python
def read_csv(file):
    try:
        return pd.read_csv('data/atendimentos/{}'.format(file))
    except:
        return pd.read_csv('data/atendimentos/{}'.format(file), encoding='latin1')
```

```python
df_list = [read_csv(file) for file in os.listdir('data/atendimentos') if 'csv' in file]
```

```python
df = pd.concat(df_list)
```
